```python
# Task 1: Dapatkan daftar semua Perusahaan dan urutkan berdasarkan Nama Perusahaan dalam urutan terbalik.
def sort_companies_by_name(company_detail_list):
    sorted_companies = sorted(company_detail_list, key=lambda x: x['name'], reverse=True)
    return [{'name': company['name']} for company in sorted_companies]
```

```python
# Task 2: Cetak semua nilai Domain di setiap perusahaan
def print_company_domains(company_detail_list):
    for company in company_detail_list:
        print(f"{company['name']}: {company['domain']} ({company['country']})")
```

```python
# Task 4: Karyawan beserta negara asal perusahaan
def employees_with_company_country(employee_detail_list, company_detail_list):
    result = []
    company_country_dict = {company['name']: company['country'] for company in company_detail_list}
    
    for employee in employee_detail_list:
        employee_info = {
            'full_name': employee['full_name'],
            'company': employee['company'],
            'country': company_country_dict[employee['company']]
        }
        result.append(employee_info)
    return result
```

```python
# Task 5: Perusahaan beserta jumlah karyawan berdasarkan kebangsaan
def company_employee_nationality(employee_detail_list):
    nationality_count = {}
    for employee in employee_detail_list:
        company_name = employee['company']
        nationality = employee['nationality']
        
        if company_name not in nationality_count:
            nationality_count[company_name] = {}
        
        if nationality not in nationality_count[company_name]:
            nationality_count[company_name][nationality] = 0
        
        nationality_count[company_name][nationality] += 1
    
    result = [{'company': company, 'employee_nationality': nationalities} for company, nationalities in nationality_count.items()]
    return result
```
